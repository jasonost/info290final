Drop rows with sketchy data

```python
events = events[(events.activityType != 'DISCUSS') & (events.eventType != 'OPENED')]
```

```python
submits = events[(events.activityType == 'ASSIGNMENT') & 
                 (events.eventType == 'CLOSED') & 
                 (events.anchorValue.apply(lambda x: pd.isnull(x))) &
                 (events.scorePercent.apply(lambda x: pd.notnull(x))) &
                 ((events.scorePercent + events.timeMinutes).apply(lambda x: x > 0))][['userId','updatedAt','topicTitle','timeMinutes','scorePercent']]
```

```python
submits2 = submits.set_index(['userId','topicTitle'])
```

```python
# procedure to determine average improvement from first submission to last submission on each assignment
def improve(uid):
    '''
    df: pandas dataframe with submission data
    uid: string, userId
    '''
    ua = submits[submits.userId == uid].groupby('topicTitle')
    topics = ua.topicTitle.nunique().index
    diffs = []
    for t in topics:
        scores = submits2.xs(uid).xs(t)
        if isinstance(scores,pd.core.series.Series):
            diffs.append(0)
        else:
            vals = scores.sort('updatedAt').scorePercent.values
            diffs.append(vals[-1] - vals[0])
    return float(sum(diffs)) / len(diffs)
```

```python
good_users = list(grades['userId'].values)
assign_behav = {u: {'total_submit': submits.groupby('userId').topicTitle.count().loc[u],
                    'uniq_assign': submits.groupby('userId').topicTitle.nunique().loc[u],
                    'avg_diff': improve(u)
                    }
                for u in good_users if u in submits.userId.values}
```
